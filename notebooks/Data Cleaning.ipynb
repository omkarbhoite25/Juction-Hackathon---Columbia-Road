```
from datetime import date, datetime
def generateTime(stringg):
  timeList =  []
  for s in stringg:
    date_raw = str(s)
    dt = datetime(
        year=int(date_raw[:4]),
        month=int(date_raw[4:6]),
        day=int(date_raw[6:8]),
     )
    timeList.append(int(dt.timestamp()))
  print(len(timeList))
  return timeList
```

#CPT

```
import pandas as pd
finalDB = pd.read_csv('https://raw.githubusercontent.com/omkarbhoite25/Juction-Hackathon-Columbia-Road/main/datasets/transaction_best_costumers.csv')
```

```
finalDB['CUST_CODE'].value_counts()
```

```
def predictCustomer(cust):
  smallDB = finalDB.copy()
  smallDB = smallDB[smallDB['CUST_CODE'] == cust]
  products = set(smallDB['PROD_CODE_10'])
  traindata =  list()
  testdata = list()
  validation = list()
  products_valid = list()
  for p in products:
    sub_df = smallDB[smallDB['PROD_CODE_10'] == p]
    #print(smallDB[smallDB['PROD_CODE_10'] == p].head())
    timestamps = list(sub_df['TIMESTAMP'])
    if len(timestamps) >= 3:
      traindata.append(timestamps[:-2])
      tmp = list()
      tmp.append(timestamps[-1])
      testdata.append(tmp)
      validation.append(timestamps[:-1])
      products_valid.append(p)
    #else:
      #print("Not Enough data")

  return traindata,testdata,validation,products_valid

```

```
from cpt.cpt import Cpt
from scipy import *
customs = set(finalDB['CUST_CODE'].tolist())

total = 0
bad = 0

for c in customs:
  train,test,val,prod = predictCustomer(c)
  #train = convolve2d(train)
  model = Cpt()  
  model.fit(train)
  out = model.predict(val)

  if len(out) != len(test) or len(out) != len(prod):
    print("FAILED")
    

  for i in range(len(prod)):
    if out[i] > (test[i][0]-400000) and out[i] < (test[i][0]+400000):
      print("SUCCESS for -> " + c + " predited: " + str(out[i]) + " on " + prod[i])
      show_db = (finalDB[finalDB['CUST_CODE'] == c])
      show_db = show_db[show_db['PROD_CODE_10']==prod[i]]
      print(show_db[['TIMESTAMP', 'PROD_CODE', 'PROD_CODE_10','CUST_CODE', 'BASKET_ID']])
      print(' ')
    else:
      bad += 1
    total += 1

print("ACCURACY:  " + str(1-bad/total))

  

# Output: ['me', 'is']
```

#BAYESIAN REGRESSION

```
import pandas as pd
finalDB = pd.read_csv('https://raw.githubusercontent.com/omkarbhoite25/Juction-Hackathon-Columbia-Road/main/datasets/transaction_best_costumers.csv')
```

```
finalDB.info()
```

```
customs = set(finalDB['CUST_CODE'])
timestamp = finalDB['TIMESTAMP'].tolist()
category = finalDB['QUANTITY'].tolist()
future_df = {}

for c in customs:
  future_df[c] = list()

print(future_df.keys())
count = 0
for t in timestamp:
  smallDB = finalDB[finalDB['TIMESTAMP']==t]
  shortcust = smallDB['CUST_CODE'].tolist()
  for key in customs:
    if key in shortcust:
      future_df[key].append(1)
    else:
      future_df[key].append(0)


future_df['TIMESTAMP'] = timestamp
future_df['CATEGORY'] = category
print(len(future_df['TIMESTAMP']))
print(len(future_df[list(future_df.keys())[0]]))
bayesian_df = pd.DataFrame(future_df)
bayesian_df.head()
```

```
pos=0
for c in customs:
  bayesian_df[c] = bayesian_df[c]*pow(2,pos)
  pos += 1
bayesian_df['ENCODE'] = bayesian_df.sum(axis=1)-bayesian_df['TIMESTAMP']

bayesian_df.head()
```

```
bayesian_df = bayesian_df[['TIMESTAMP', 'ENCODE', 'CATEGORY']]

from sklearn.model_selection import train_test_split# df is features and labels are the targets 

# Split by putting 25% in the testing set
X_train, X_test, y_train, y_test = train_test_split(bayesian_df[['TIMESTAMP','CATEGORY']],bayesian_df['ENCODE'],
                                                   test_size = 0.1,
                                                    random_state=42)
```

```

import numpy as np 
import matplotlib.pyplot as plt 
  
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score 
import statsmodels.api as sm 
```

```
regression_model = LinearRegression() 
  
# Fit the data(train the model) 
regression_model.fit(X_train, y_train) 
  
# Predict 
y_predicted = regression_model.predict(X_test) 
  
# model evaluation 
mse=mean_squared_error(y_test,y_predicted)
print(y_test)
print(y_predicted)
  
rmse = np.sqrt(mean_squared_error(y_test, y_predicted)) 
r2 = r2_score(y_test, y_predicted) 
  
# printing values 
print('Slope:' ,regression_model.coef_) 
print('Intercept:', regression_model.intercept_) 
print('MSE:',mse) 
print('Root mean squared error: ', rmse) 
print('R2 score: ', r2) 
```
